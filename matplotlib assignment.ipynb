```python
"""
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Ans.A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional graph. Each dot represents a 
    single data observation, and the position of the dot on the graph is determined by the values of two variables, one on the x-axis and the other
    on the y-axis. Scatter plots are useful for visualizing the relationship between two continuous variables and are often used to identify
    patterns, trends, clusters, or outliers in the data.
"""
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y, c='blue', marker='o')

plt.title('Scatter Plot of X vs. Y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.grid(True)
plt.show()

```

```python
"""
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans.The subplot() function in Matplotlib is used to create multiple subplots (smaller individual plots) within a single figure. It allows you to 
    display multiple plots in a grid-like arrangement, making it easier to visualize and compare different datasets or aspects of your data in one 
    figure. Each subplot is identified by its row and column position within the grid.
"""
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x, y4)
plt.title('Line Plot 4')

plt.tight_layout()

plt.show()
```

```python
"""
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Ans.A bar plot, also known as a bar chart or bar graph, is a type of data visualization that displays data using rectangular bars or columns. In a 
    bar plot, the length or height of each bar is proportional to the data it represents. Bar plots are used to compare and represent discrete 
    categories of data, making it easy to visualize and compare values among different categories or groups.
    
    Bar plots are used for the following purposes:
    Comparing Categories
    Displaying Categorical Data
    Showing Trends
    Highlighting Differences
"""

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(8, 4))
plt.bar(company, profit, color='black')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Company Profit Comparison (Bar Plot)')
plt.show()

plt.figure(figsize=(8, 4))
plt.barh(company, profit, color='red')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Company Profit Comparison (Horizontal Bar Plot)')
plt.gca().invert_yaxis()  
plt.show()
```

```python
"""
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

Ans.A box plot, also known as a box-and-whisker plot, is a data visualization technique used to display the distribution of a dataset. It provides 
    a summary of the dataset's central tendency, spread, and any potential outliers. Box plots are particularly useful for comparing the 
    distributions of multiple datasets or variables.
    
    Here's why box plots are used:
    Summary of Data Distribution
    identification of Outliers
    Comparison of Distributions
"""
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot of Two Datasets')
plt.grid(True)
plt.show()
```
