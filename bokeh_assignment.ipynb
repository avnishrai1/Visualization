```python
"""
Q1. How can you create a Bokeh plot using Python code?
Ans.Import Bokeh: First, you need to import the Bokeh library.

    Create a Figure: Create a figure object that will serve as the canvas for your plot. You can customize various aspects of the plot, such as its size, title,
    and axis labels, by passing parameters to the figure function.

    Add Data to the Plot: Use Bokeh's glyph functions (e.g., circle, line, square, etc.) to add data to the plot. For example, if you want to create a scatter plot,
    you can use the circle glyph function

    Customize the Plot: You can further customize the plot by adjusting various attributes and properties of the figure object, such as colors, line styles, and more.

    Show the Plot: Finally, use the show function to display the plot in your Python environment (e.g., Jupyter Notebook, Python script, or interactive Python session).
"""
import bokeh.io
import bokeh.plotting
from bokeh.plotting import figure, output_file, show
bokeh.io.output_notebook()

p = figure(title="My Bokeh Plot", width=800, height=400, x_axis_label="X-axis", y_axis_label="Y-axis")

x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

p.circle(x, y, size=10, color="blue")

show(p)
```

```python
"""
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
Ans.In Bokeh, glyphs are visual shapes or markers that are used to represent data points on a plot. Glyphs are the building blocks of Bokeh visualizations and can include
    shapes like circles, squares, triangles, lines, and more. You can customize these glyphs by mapping data attributes to their properties, such as position, size, color, and shape.
"""
import bokeh.io
import bokeh.plotting
from bokeh.plotting import figure, output_file, show
bokeh.io.output_notebook()

x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]

p = figure(title="Scatter Plot Example", width=800, height=400)

p.circle(x, y, size=10, color="blue", alpha=0.5)

p.xaxis.axis_label = "X-axis"
p.yaxis.axis_label = "Y-axis"
p.title.text_color = "green"
p.title.text_font_size = "16px"

show(p)
```
