```python
"""
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Ans.Seaborn is a versatile library for creating various types of statistical plots. Here are five common plots and their typical 
    use cases:
    
    Scatter Plot (sns.scatterplot()):  Scatter plots are used to visualize the relationship between two continuous variables. 
    Each data point is represented as a dot on the plot, making it useful for identifying patterns, clusters, or trends in data.
    
    Histogram (sns.histplot()): Histograms are used to display the distribution of a single variable. They show the frequency 
    or count of data points within specific bins or intervals, helping you understand the data's underlying structure.
    
    Bar Plot (sns.barplot()): Bar plots are ideal for displaying categorical data and comparing values across different 
    categories. They are useful for visualizing counts, sums, or averages of data for different groups or categories
    
    Box Plot (sns.boxplot()):  Box plots, also known as box-and-whisker plots, are used to visualize the distribution of a dataset's 
    summary statistics, such as the median, quartiles, and potential outliers. They are helpful for identifying data skewness and spread.
    
    Violin Plot (sns.violinplot()): Violin plots combine a box plot with a kernel density estimate, providing a comprehensive view of the 
    data's distribution.
"""
import seaborn as sns
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 9, 11]
```

```python
# Create a scatter plot
sns.scatterplot(x=x, y=y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.show()
```

```python
# Sample data
data = [3, 3, 5, 5, 5, 7, 7, 9, 9, 11]

# Create a histogram
sns.histplot(data)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

```python
# Sample data
categories = ['Category A', 'Category B', 'Category C']
values = [20, 35, 15]

# Create a bar plot
sns.barplot(x=categories, y=values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()
```

```python
# Sample data
data = [20, 25, 30, 35, 40, 45, 50, 55, 60]

# Create a box plot
sns.boxplot(data=data)
plt.ylabel('Value')
plt.title('Box Plot Example')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({'Category': ['Category A', 'Category B', 'Category C'] * 3,
                     'Values': [30, 45, 20, 50, 35, 25, 40, 60, 55]})

sns.violinplot(x='Category', y='Values', data=data)
plt.xlabel('Categories')
plt.ylabel('Values')
```

```python
"""
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for 
    different events and regions.
"""
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")

sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)

plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Line Plot of FMRI Data")
```

```python
"""
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and 
    y = 'fare'.
    Note: pclass, age, and fare are columns in the titanic dataset.
"""
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.title('Box Plot of Age by Pclass')

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.title('Box Plot of Fare by Pclass')
```

```python
"""
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column 
    of the diamonds dataset.
"""
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

sns.histplot(data=diamonds, x='price', hue='cut')

plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Diamond Prices by Cut')
```

```python
"""
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
"""
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

sns.pairplot(hue = 'species', data = iris)
```

```python
"""
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
"""
import seaborn as sns
import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")

flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

plt.figure(figsize=(16, 10))
sns.heatmap(data=flights_pivot, fmt='d', annot = True)

plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Count Heatmap by Year and Month")
```
